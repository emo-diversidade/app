```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
import seaborn as sns

data = []

# ... (Dados dos participantes 1 e 2)

# Funções para gerar dados com distribuições diferentes por tipo de relacionamento
def gerar_compersao(tipo_relacionamento):
    if tipo_relacionamento == 'Swing':
        return np.random.choice([1, 2, 3, 4, 5], p=[0.1, 0.1, 0.2, 0.3, 0.3])  # Mais alto
    elif tipo_relacionamento == 'Aberto':
        return np.random.choice([1, 2, 3, 4, 5], p=[0.25, 0.25, 0.2, 0.15, 0.15])  # Mais baixo
    elif tipo_relacionamento == 'Poliamor':
        return np.random.choice([1, 2, 3, 4, 5], p=[0.05, 0.15, 0.3, 0.3, 0.2])  # Médio
    else:
        return np.random.randint(1, 6)

def gerar_satisfacao_relacional(tipo_relacionamento):
    if tipo_relacionamento == 'Swing':
        return np.random.choice([1, 2, 3, 4, 5], p=[0.1, 0.1, 0.2, 0.3, 0.3])  # Mais alto
    elif tipo_relacionamento == 'Aberto':
        return np.random.choice([1, 2, 3, 4, 5], p=[0.3, 0.25, 0.2, 0.15, 0.1])  # Mais baixo
    elif tipo_relacionamento == 'Poliamor':
        return np.random.choice([1, 2, 3, 4, 5], p=[0.05, 0.1, 0.25, 0.35, 0.25])  # Médio
    else:
        return np.random.randint(1, 6)

# ... (Outras funções para gerar variáveis com distribuições diferentes, se necessário)


# Gerando dados para os participantes 3 a 40
for participante in range(3, 41):
    grupo = 'Experimental' if participante % 2 != 0 else 'Controle'
    tipo_relacionamento = random.choice(['Swing', 'Poliamor', 'Aberto'])
    tempo_relacionamento_meses = np.random.randint(40, 121)
    # ... (outras variáveis)

    for momento_medicao in range(1, 4):
        compersao = gerar_compersao(tipo_relacionamento)
        satisfacao_relacional = gerar_satisfacao_relacional(tipo_relacionamento)
        # ... (geração de outras variáveis)

        data.append([participante, grupo, tipo_relacionamento, tempo_relacionamento_meses, momento_medicao, compersao, intensidade_compersao, paresia, intensidade_paresia, ciume, satisfacao_relacional, comunicacao, autoestima, duracao_relacionamento_secundario_meses, qualidade_relacionamento_secundario, numero_parceiros_secundarios, estilo_comunicacao, estilo_resolucao_conflitos, frequencia_comunicacao, satisfacao_sexual_individual, satisfacao_sexual_casal, suporte_parceiro_nm, gestao_tempo_nm, historico_relacionamentos, uso_aplicativos_encontros, acordos_regras_nm, percepcao_stigma, autoconsciencia_emocional, compaixao_pelo_parceiro, investimento_relacionamento_principal, motivacao_nm, meta_acordos, conhecimento_nm, frequencia_negociacao_acordos, terapia_casal, divisao_tarefas, independencia_financeira, espaco_individual, influencia_familia_amigos, satisfacao_vida, equilibrio_vida_pessoal_relacionamento, tempo_dedicado_relacionamento_principal, qualidade_tempo_relacionamento_principal, hobbies_interesses_compartilhados, individualidade_parceiro, seguranca_estabilidade_relacionamento, criatividade_sexual, intimidade_emocional])

# Criando o DataFrame
columns = ['participante', 'grupo', 'tipo_relacionamento', 'tempo_relacionamento_meses', 'momento_medicao', 'compersao', 'intensidade_compersao', 'paresia', 'intensidade_paresia', 'ciume', 'satisfacao_relacional', 'comunicacao', 'autoestima', 'duracao_relacionamento_secundario_meses', 'qualidade_relacionamento_secundario', 'numero_parceiros_secundarios', 'estilo_comunicacao', 'estilo_resolucao_conflitos', 'frequencia_comunicacao', 'satisfacao_sexual_individual', 'satisfacao_sexual_casal', 'suporte_parceiro_nm', 'gestao_tempo_nm', 'historico_relacionamentos', 'uso_aplicativos_encontros', 'acordos_regras_nm', 'percepcao_stigma', 'autoconsciencia_emocional', 'compaixao_pelo_parceiro', 'investimento_relacionamento_principal', 'motivacao_nm', 'meta_acordos', 'conhecimento_nm', 'frequencia_negociacao_acordos', 'terapia_casal', 'divisao_tarefas', 'independencia_financeira', 'espaco_individual', 'influencia_familia_amigos', 'satisfacao_vida', 'equilibrio_vida_pessoal_relacionamento', 'tempo_dedicado_relacionamento_principal', 'qualidade_tempo_relacionamento_principal', 'hobbies_interesses_compartilhados', 'individualidade_parceiro', 'seguranca_estabilidade_relacionamento', 'criatividade_sexual', 'intimidade_emocional']
df = pd.DataFrame(data, columns=columns)

# Pré-processamento para a visualização (convertendo para numérico)
for col in ['intensidade_compersao', 'intensidade_paresia', 'qualidade_relacionamento_secundario', 'satisfacao_sexual_individual', 'satisfacao_sexual_casal', 'suporte_parceiro_nm', 'percepcao_stigma', 'autoconsciencia_emocional', 'compaixao_pelo_parceiro', 'investimento_relacionamento_principal', 'conhecimento_nm', 'tempo_dedicado_relacionamento_principal', 'qualidade_tempo_relacionamento_principal', 'hobbies_interesses_compartilhados', 'individualidade_parceiro', 'seguranca_estabilidade_relacionamento', 'criatividade_sexual', 'intimidade_emocional', 'divisao_tarefas', 'independencia_financeira', 'espaco_individual', 'influencia_familia_amigos', 'satisfacao_vida', 'equilibrio_vida_pessoal_relacionamento']:
    df[col] = df[col].astype('category').cat.codes

# Visualização: Gráfico de dispersão com cores vibrantes
plt.figure(figsize=(10, 6))
sns.scatterplot(x='compersao', y='satisfacao_relacional', hue='tipo_relacionamento', data=df, palette='viridis', s=100, alpha=0.7)
plt.title('Relação entre Compersão e Satisfação Relacional (Colorido por Tipo de Relacionamento)')
plt.xlabel('Compersão')
plt.ylabel('Satisfação Relacional')
plt.show()


df.to_csv('synthetic_data_with_relationship_quality.csv', index=False)
print(df)
```
