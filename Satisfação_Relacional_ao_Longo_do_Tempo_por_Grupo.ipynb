```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Dados
data = []

# Participantes 1 e 2 (exemplos específicos)
data.extend([
    [1, 'Experimental', 'Swing', 120, 1, 4, 3, 2, 4, 5, 4, 2, 1, 4, 4, 4, 1, 5],
    [1, 'Experimental', 'Swing', 120, 2, 5, 4, 1, 5, 5, 5, 1, 1, 5, 5, 5, 1, 5],
    [1, 'Experimental', 'Swing', 120, 3, 5, 5, 1, 5, 5, 5, 1, 1, 5, 5, 5, 1, 5],
    [2, 'Controle', 'Swing', 80, 1, 3, 2, 3, 3, 4, 3, 3, 2, 3, 3, 3, 2, 4],
    [2, 'Controle', 'Swing', 80, 2, 3, 2, 3, 3, 4, 3, 3, 2, 3, 3, 3, 2, 4],
    [2, 'Controle', 'Swing', 80, 3, 3, 2, 3, 3, 4, 3, 3, 2, 3, 3, 3, 2, 4]
])

# Participantes 3 a 40 (dados sintéticos)
for participant in range(3, 41):
    grupo = 'Experimental' if participant % 2 != 0 else 'Controle'
    tipo_relacionamento = 'Swing'  # Ou pode ser aleatório
    tempo_relacionamento = np.random.randint(40, 121)
    for momento in range(1, 4):
        row = [participant, grupo, tipo_relacionamento, tempo_relacionamento, momento]
        row.extend([np.random.randint(1, 6) for _ in range(13)])  # Gera 13 valores aleatórios (corrigido)
        data.append(row)

columns = ['participante', 'grupo', 'tipo_relacionamento', 'tempo_relacionamento_meses', 'momento_medicao', 'compersao', 'paresia', 'ciume', 'satisfacao_relacional', 'comunicacao', 'autoestima', 'pensamentos_intrusivos', 'ansiedade_social', 'suporte_social', 'empatia', 'regulacao_emocional', 'comportamentos_evitativos', 'flexibilidade_cognitiva']
df = pd.DataFrame(data, columns=columns)


# Gráfico de barras
plt.figure(figsize=(12, 6))
sns.barplot(x='momento_medicao', y='satisfacao_relacional', hue='grupo', data=df, palette='viridis')
plt.title('Satisfação Relacional ao Longo do Tempo por Grupo')
plt.xlabel('Momento da Medição')
plt.ylabel('Satisfação Relacional')
plt.show()

df.to_csv('synthetic_data_with_groups_corrected.csv', index=False)
print(df)
```
